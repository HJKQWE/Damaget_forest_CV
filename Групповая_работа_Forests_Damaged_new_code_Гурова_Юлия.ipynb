## Источники

Для обучения модели для задачи классификации мы использовали изображения, находящиеся в свободном доступе, снятые со спутников, не имеющие гео-метки. Но путем дальнейшего усовершенствования и использования других моделей мы ожидаем получить уверенный результат детекции. Далее мы планируем разработать матмодель для подсчета выбросов СО2, согласно ГОСТу, а так же рассчитать сокращение эмиссии кислорода погибшим биомом.

Величина выбросов поллютантов в атмосферу зависит от типа лесного пожара, запаса и типа лесного горючего материала и его влагосодержания, типа лесных фитоценозов, метеоусловий, рельефа местности и ряда других условий. Для оценки сверху величины выбросов поллютантов в атмосферу при возгорании лесных фитоценозов целесообразно иметь так называемую максимальную модель леса.

Под максимальной моделью леса будем понимать гипотетический лесной массив, в котором запас лесных горючих материалов для каждого из ярусов леса максимален, а их влагосодержание минимально. Очевидно, что при горении такого лесного массива температура горения и интенсивность пожара (количество тепловой энергии, выделившейся на единицу длины кромки пожара) будут максимальны, чем и объясняется название - максимальная модель леса. Как правило, при лесных пожарах сгорают тонкие (до 7 мм в диаметре) веточки, хвоинки (в хвойных лесах) и листья (в лиственных). На основании данных о запасе лесных горючих материалов для различных ярусов леса и теплотворной способности, а также вышеизложенного, в качестве максимальной модели хвойного леса предлагается следующая совокупность ярусов гипотетического леса:

1. Ярус мхов, лишайников с включениями из опавших хвоинок и тонких веточек (нулевой слой). Высота его составляет 15 см, плотность r=20 кг/м3, запас ЛГМ m0=3,0 кг/м2, теплотворная способность q= 19446 кДж/кг, влагосодержание W=8%.

2. Первый ярус леса - травы и кустарники. Высота слоя 2 м, плотность ЛГМ в этом слое rс=0.8 кг/м3, запас ЛГМ в первом ярусе m0=1.6 кг/м2, теплотворная способность q=17808 кДж/кг, влагосодержание W=80%.

3. Второй ярус леса - подрост - представляет собой совокупность деревьев высотой до 6 м. Высота слоя h2 = 6 м, толщина D2 = 5 м, то есть нижняя граница полога подроста (полог подроста - совокупность крон молодых деревьев) находится в первом ярусе леса (первый и второй ярусы частично перекрываются). Плотность слоя rс = 0.4 кг/м3, запас ЛГМ m0 = 2 кг/м2, теплотворная способность q = 21949 кДж/кг, влагосодержание W = 80%.

4. Третий ярус леса - совокупность крон деревьев. Высота верхней границы полога леса  = 22 м, высота нижней - h3 = 5 м, толщина слоя D =  - h3 = 17 м, то есть второй и третий ярусы леса частично перекрываются. Плотность слоя ЛГМ в пологе леса rс = 0.3 кг/м3, запас ЛГМ m3 = 5,1 кг/м2, теплотворная способность q = 21949 кДж/кг, содержание воды W = 90%.

Используя лесотаксационные описания лесных массивов, информацию о вертикально-фракционной структуре леса, данные о запасе ЛГМ, можно построить максимальные модели леса для каждого региона страны.

# <b>Создание датасетов для нейронной сети</b>

Для создания датасетов мы будем исопльзовать инструмент библиотеки `keras` `image_dataset_from_directory`, который рекомендован для написания всех новых кодов.

## `image_dataset_from_directory`

Для удобства выведем гиперапараметры `image_dataset_from_directory` в отдельные переменные.

````python
batch_size = 16
n = 250
img_height, img_width = (n, n)
label_mode = 'binary'
````

Создадим тренировочный датасет посредством метода `flow_from_directory` библиотеки `keras`. Так как мы решаем задачу бинарной классификации, то необходимо указать `label_mode='binary'`. Также сразу укажем желаемое разрешение, к которому будут приведены фотографии в датасете.

````python
train_ds = tf.keras.utils.image_dataset_from_directory(
    base_dir,
    label_mode=label_mode,
    validation_split=0.2,
    subset='training',
    seed=12,
    image_size=(img_height, img_width),
    batch_size=batch_size)
````

Проделаем ту же операцию для валидационной части.

````python
val_ds = tf.keras.utils.image_dataset_from_directory(
    base_dir,
    label_mode=label_mode,
    validation_split=0.2,
    subset='validation',
    seed=12,
    image_size=(img_height, img_width),
    batch_size=batch_size)
````

# <b>Обработка данных</b>

## Rescale

### Создание слоя Rescale

Создадим слой, который приведет значения массивов изображений к стандартному виду (от 0 до 1). Это стандартная операция по подготовке изображений для обучения нейросети. Так как даный слой будет первый в нашей сети, укажем `input_shape=(img_height, img_width, 3)`.

````python
rescale = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255,
                              input_shape=(img_height, img_width, 3),
                              name='Rescaling')
])
````

### Визуализация работы Rescale

Проверим качество работы слоя на фотографии, которую мы использовали выше.

````python
from tensorflow.python.ops.numpy_ops import np_config
np_config.enable_numpy_behavior()

# увеличим размерность изображения до 4, так иначе оно не пройдет
# корректно через слой
result = rescale(np.expand_dims(one_pic, axis=0))
# вернем результату размерность 3 для отображения
result = result.reshape(one_pic.shape)
plt.axis('off')
plt.imshow(result);

# print("Picture's shape:", result.shape)
print("Миниммальное и максимальное значение пикселей:", result.numpy().min(), result.numpy().max())
````

Как видим, слой отработал ожидаемым образом, значения пикселей приведены к диапазону от 0 до 1.

## Аугментация данных

Следующий шаг - аугментация данных. Так как наш датасет создан из сравнительно небольшого набора данных, а именно из нарезанных спутниковых снимков, то для избежания переобучения мы воспользуемся возможностями библиотеки `keras` по искусственному расширению количества фотографий через внешнее изменение имеющихся. Создадим последовательность слоев `RandomFlip` (зеркальное отображение изображения в случайном порядке), `RandomRotation` (поворот изображения по часовой стрелке на случайный угол), `RandomZoom` (увеличение и уменьшение изображения в случайном порядке), а также `RandomContrast`. Это не все воможности аугментации, представленные в библиотеке `TensorFlow`, но на данном этапе мы остановимся на этих методах.

### Создание последовательности для агументации данных

````python
import tensorflow as tf

data_aug = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal_and_vertical', name='RandomFlip'),
    tf.keras.layers.RandomRotation(factor=1, fill_mode='reflect',
                                   name='RandomRotation'),
    tf.keras.layers.RandomZoom(height_factor=(-0.7, 0.4), name='RandomZoom'),
    tf.keras.layers.RandomContrast(factor=(0.3, 1), name='RandomContrast')
])
````

### Визуалиация работы слоя data_augmentation

Как и в случае с созданием датасета, проверим на уже использованном изображении из тренировочного датасета, как работает последовательность по аугментации данных. Если следующий блок с кодом запустить несколько раз, то мы увидим, как меняется изображение, проходя через вышеописанные слои.

> _**Note**: В версии tf 2.11.0 при выполнении нижеследующего кода выводятся предупреждения, решить которые программными средствами пока не удается, поэтому мы рекомендуем использовать версию 2.8.3, в которой данный код работает без ошибок (см. раздел с техническими требованиями к проекту)._

````python
from tensorflow.python.ops.numpy_ops import np_config
np_config.enable_numpy_behavior()

# увеличим размерность изображения до 4, так иначе оно не пройдет
# корректно через последовательность слоев
result_for_aug = np.expand_dims(result, axis=0)
augmented = data_aug(result_for_aug)
print(augmented.shape)

plt.axis('off')

# вернем результату размерность 3 для отображения
plt.imshow(augmented.reshape(result.shape));
````

Итак, обе последовательности успешно справляются с поставленными перед ними задачами. Переходим к созданию и обучению моделей.

## Autotune

Также запустим код, который позволит оптимизировать использование памяти компьютера с целью не допустить ее перегрузки в процессе обучения - `cache` обеспечивает кэширование данных для ипользования в последующих эпохах. `prefetch`, в свою очередь, обеспечит ускорение процесса обучения за счет наложения подготовки данных к следующей эпохе во время вычислений на этапе валидации текущей эпохи.

# <b>Модели нейронных сетей</b>

Протестируем две модели с одними настройками основных гиперпараметров для сравнения их работы, отличие моделей - в слое аугментации. В первой модели этот слой отсутствует.

Вынесем общие гиперпараметры в отдельные переменные, это будут:
- learning rate для оптимизатора,
- epochs - количество итераций при обучении, 
- n_1, n_2 - количество нейронов для каждого слоя.

````python
learning_rate = 0.001
epochs = 30
n_1 = 512
n_2 = 512
````

## Модель #1

Первая модель нейронной сети для классификации изображений выгоревших территорий и обычных будет работать с изображениями без аугментации, то есть только на исходных изображениях датасета. Данная модель состоит из следующих трех блоков:

- блок по нормализации массива (rescale),
- блок слоёв свертки (Conv2D) и сжатия (MaxPooling2D) изображений,
- непосредственно нейронные слои.

Слои свертки пропускают изображения через фильтры размером 3x3 пикселя, которые позволяют машине найти наиболее важные признаки. Слои сжатия, в свою очередь, после каждой свертки уменьшают разрешение изображений в два раза. Таким образом, на вход нейронным слоям подаются очищенные от шума данные, за счет чего повышается эффективность работы алгоритма.

В качестве функции активации во всех слояих, кроме последнего, была выбрана ReLU (Rectified Linear Unit), которая обеспечивает нелинейность работы сети. Выбор функции активации является гиперпараметром, с выбором лучшей фнункции для данной модели можно экспериментировать, но на данном этапе мы остановимся на этом варианте.

В последнем слое в качестве функции активации применена сигмоида, так как мы предсказываем два класса. На выходе мы ожидаем одно значение - вероятность класса (`0` - `burned`, `1` - `regular`).

### Архитектура модели

````python
import tensorflow as tf

model_1 = tf.keras.models.Sequential([

    # preprocessing
    rescale,

    # conv1
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu',
                           input_shape=(img_height, img_width, 3),
                           name='Conv2D_1'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_1'),
    # conv2
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', name='Conv2D_2'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_2'),
    # conv3
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_3'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_3'),
    # conv4
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_4'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_4'),
    # conv5
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_5'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_5'),
    # conv6
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_6'),

    # neural layers
    tf.keras.layers.Flatten(name='Flatten'),
    tf.keras.layers.Dense(n_1, activation='relu', name='Dense_1'),
    tf.keras.layers.Dense(n_2, activation='relu', name='Dense_2'),
    tf.keras.layers.Dense(1, activation='sigmoid', name='Dense_output')

])
````

В блоке ниже можно увидеть, какие изменения происходят с изображениями по мере прохождения через сеть.

````python
model_1.summary()
````

### Компиляция

На данном этапе мы выбираем алогоритм для обучения модели, функцию потерь и метрики для отслеживания качества обучения. Эти параметры являются гиперпараметрами, поэтому с выбором конкретного набора можно поэкспериментировать. Для нашей модели мы взяли алгоритм `RMSprop`, который адаптирует `learning rate` в ходе обучения модели. Функция потерь - бинарная кросс-энтропия, стандартная в современных моделях функция потерь для бинарной классификации. В качестве метрики мы также взяли стандартную метрику `accuracy`, сбалансированность классов позволяет нам это сделать без опаски пропустить ошибки работы алгоритма.

````python
from tensorflow.keras.optimizers import RMSprop

model_1.compile(optimizer=RMSprop(learning_rate=learning_rate),
             loss=tf.keras.losses.BinaryCrossentropy(),
             metrics=['accuracy'])
````

### Обучение модели

Обучение модели будем производить на нашем тренировочном датасете, а оценку ее качества - на валидационном. Результат работы алгоритма мы будем видеть в онлайн-режиме, установив `verbose` на 1.

````python
# epochs = 20
history_1 = model_1.fit(
            train_ds,
            epochs=epochs,
            validation_data=val_ds,
            verbose=1
            )
````

### Визуализация данных по работе алгоритма

Посмотрим на то, как вели себя функция потерь и `accuracy` на тренировочном и валидационном датасетах на всех итерациях модели #1.

````python
acc = history_1.history['accuracy']
val_acc = history_1.history['val_accuracy']

loss = history_1.history['loss']
val_loss = history_1.history['val_loss']

epochs_range = range(1, epochs+1)

plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='best')
plt.title('Accuracy model #1')
plt.xlabel('epochs')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='best')
plt.title('Loss model #1')
plt.xlabel('epochs')
plt.show()

val_acc_1 = val_acc
````

Как показывает статистика, 30 итераций при данных настройках и архитектуре сети - многовато, можно вполне ограничиться 15-20 эпохами, так как переобучение начинается примерно на этом этапе.

### Предсказание модели

Проверим качество работы модели на реальных фотографиях экспертным методом. В качестве экспертов выступим мы, пропустив через модель заранее подготовленные изображения лесов и сравнив иображения с тем результатом, который выдает нам наша нейронная сеть. Буква `b` в названии файлов означает принадлежность к классу `burned`, `r` - `regular`.

````python
import numpy as np
from tensorflow.keras.utils import load_img, img_to_array
import os

# images = os.listdir('./data/test')
images = os.listdir('/content/drive/MyDrive/Sber DS/GroupAssignment2/test')  # for collab

print(images)

for i in images:
    print()
    # путь к тестовым изображениям
    # path = './data/test/' + i
    path = '/content/drive/MyDrive/Sber DS/GroupAssignment2/test/' + i  # for colab
    
    # приведение изображений к желаемому разрешению
    img = load_img(path, target_size=(img_height, img_width))
    # перевод изображения в массив
    x = img_to_array(img)
    # plt.imshow(x.astype('uint8'))
    # plt.show()

    # добавление четвертого измерения для модели
    images = np.expand_dims(x, axis=0)
    # plt.imshow(images.reshape(img_height, img_width, 3))
    # plt.show()
    # предсказание
    classes = model_1.predict(images, batch_size=10)
    print(classes[0])
    if classes[0]>0.5:
        print(i + ' has a regular forest')
    else:
        print(i + ' has a burned forest')
````

## Модель #2

Построим вторую модель нейронной сети на базе предыдущей, добавив в нее слои аугментации. Данная модель состоит из следующих блоков:

- блок обработки данных: стандартизация и аугментация данных,
- блок слоёв свертки (Conv2D) и сжатия (MaxPooling2D) изображений,

Все основные характеристики остались без изменений, мы только добавляем четыре слоя для аугментации данных.

### Архитектура модели

````python
import tensorflow as tf

model_2 = tf.keras.models.Sequential([

    # preprocessing
    rescale,
    data_aug,

    # conv1
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu',
                           input_shape=(img_height, img_width, 3),
                           name='Conv2D_1'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_1'),
    # conv2
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', name='Conv2D_2'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_2'),
    # conv3
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_3'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_3'),
    # conv4
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_4'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_4'),
    # conv5
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_5'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), name='MaxPooling2D_5'),
    # conv6
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', name='Conv2D_6'),

    # neural layers
    tf.keras.layers.Flatten(name='Flatten'),
    tf.keras.layers.Dense(n_1, activation='relu', name='Dense_1'),
    tf.keras.layers.Dense(n_2, activation='relu', name='Dense_2'),
    tf.keras.layers.Dense(1, activation='sigmoid', name='Dense_output')

])
````

# <b>Визуализация свертки изображений</b>

В завершение посмотрим, как работает с нашими изображениями сеть во время свертки. Запустив следующий код, мы проследим работу сети на каждом сверточном слое и слое сжатия, увидим, как машина обнаруживает наиболее важные признаки и с каким итогом она подходит к своему первому нейронному слою.

1. Слои свертки и сжатия:

````python
import numpy as np
import random
from tensorflow.keras.utils import load_img, img_to_array

# Define a new Model that will take an image as input, and will output
# intermediate representations for all layers in the previous model after
# the first block.
successive_outputs = [layer.output for layer in model_1.layers[2:]]
visualization_model = tf.keras.models.Model(
                                inputs=model_1.input,
                                outputs=successive_outputs)

# Prepare a random input image from the training set.
burned_img_files = [os.path.join(burned_dir, fname)
                    for fname in burned_fnames]
regular_img_files = [os.path.join(regular_dir, fname)
                     for fname in regular_fnames]
img_path = random.choice(burned_img_files + regular_img_files)

img = load_img(img_path, target_size=(img_height, img_width))
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

successive_feature_maps = visualization_model(x)
layer_names = [layer.name for layer in model_1.layers[2:]]

for layer_name, feature_map in zip(layer_names, successive_feature_maps):
    if len(feature_map.shape) == 4:
        # number of features in feature map
        n_features = feature_map.shape[-1]
        # The feature map has shape (1, size, size, n_features)
        size = feature_map.shape[1]
        # Tile the images in this matrix
        display_grid = np.zeros((size, size*n_features))
        
        for i in range(n_features):
            x = feature_map[0, :, :, i]
            x -= x.numpy().mean()
            x /= x.numpy().std() + 0.000001
            x *= 64
            x += 128
            x = np.clip(x, 0, 255).astype('uint8')
            
            # Tile each filter into this big horizontal grid
            display_grid[:, i*size : (i+1)*size] = x
            
        # print(display_grid.shape)
        scale = 20. / n_features
        plt.figure(figsize=(scale * n_features, scale))
        plt.title(layer_name)
        plt.grid(False)
        plt.imshow(display_grid, aspect='auto', cmap='viridis')
````
